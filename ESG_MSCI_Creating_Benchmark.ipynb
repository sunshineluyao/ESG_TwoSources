# Step 2: Calculating Industry Adjusted BMR Quarterly

### 1. calculating Longterm_BV, Longterm_MV, and Longterm BMV


```
df['Longterm_BV']= (df['BV_Y1']+df['BV_Y2']+df['BV_Y3']+df['BV_Y4']+df['BV_Y5'])/5
df['Longterm_MV']= (df['MV_Y1']+df['MV_Y2']+df['MV_Y3']+df['MV_Y4']+df['MV_Y5'])/5
df['Longterm_BMR']=df['Longterm_BV']/df['Longterm_MV']
df.head()
```

###2. Caculating Industry_Ave_BMR

```
df['SIC_2']=df['SIC_Code'].apply(lambda x: int(x/100))
df.head()
```

```
df_temp=df.groupby(['SIC_2'], as_index=False)['Longterm_BMR'].mean()
df_temp.columns = ['SIC_2','Industry_Ave_BMR']
df_temp
```

```
df=df.merge(df_temp, on=['Month','SIC_2'])
df.head()
```

### 3. Calculate BMR Quarterly

```
df['BMR_Quarterly']=df['BookValue_Quarterly']/df['MarketValue_Quarterly']
df.head()
```

### Calculating Industry_Adjusted_BMR_Quarterly 

```
df['Industry_Adjusted_BMR_Quarterly']=df['BMR_Quarterly']-df['Industry_Ave_BMR']
df.head()
```

# Step 3: Creating Size_G, BMR_G, and Mom_G

### For each month, Generate the categories by quintile cut of size

```
Month_List = df['Month'].unique().tolist()
Month_List
```

```
df_ =pd.DataFrame()
for i in Month_List:
  df_temp = df[df.Month==i]
  df_temp['Size_G']=pd.qcut(df_temp.Size,5,labels=np.arange(1,6,1))
  df_ =df_.append(df_temp)
df_.head()
```

```
df=df_
```

### For each month, Generate the categories by quintile of BM Ratio within each Size group.

```
df_ =pd.DataFrame()
for i in Month_List:
  df_temp = df[df.Month==i]
  df_temp['BMR_G'] = df_temp.groupby(['Size_G'])['Industry_Adjusted_BMR_Quarterly'].transform(lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
  df_ =df_.append(df_temp)
df_.head()
```

```
df =df_
```
